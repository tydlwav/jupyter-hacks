# Plotting as code runs

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import time
from IPython.display import display, clear_output

x = np.linspace(0,2*np.pi)
y = 2*np.sin(x)

max_x = np.max(x)
max_y = np.max(y)
min_x = np.min(x)
min_y = np.min(y)
margin_x = (max_x - min_x)*0.1
margin_y = (max_y - min_y)*0.1
left_x = min_x - margin_x
right_x = max_x + margin_x
left_y = min_y - margin_y
right_y = max_y + margin_y

fig = plt.figure(figsize=(12, 4))
sub = fig.add_subplot(1,1,1)

for i in range(len(x)):
    sub.cla()
    axes = plt.gca()
    axes.set_xlim(left_x, right_x)
    axes.set_ylim(left_y, right_y)
    sub.plot(x[:i],y[:i])
    display(fig)
    clear_output(wait=True)

plt.show() # Optional if you have inline
```
